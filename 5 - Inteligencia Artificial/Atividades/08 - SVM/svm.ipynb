```python
import pandas as pd

# Carrega o arquivo
arquivo = pd.read_csv('./data/treino_5x5.csv')
```

```python
# Mostra as 5 primeiras linhas
arquivo.head()
```

```python
# Definindo as variáveis preditoras e a variável alvo
y = arquivo['classe']
x = arquivo.drop('classe', axis = 1)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler

# Normaliza os dados
normalizador = MinMaxScaler(feature_range=(0, 1))
x_normalizado = normalizador.fit_transform(x)

# Cria o modelo
modelo = SVC()

# Cria o modelo de validação cruzada
kfold = StratifiedKFold(n_splits=3)

# Executa a validação cruzada
resultado = cross_val_score(modelo, x_normalizado, y, cv=kfold, n_jobs=-1)

# Imprime o resultado
print(resultado.mean())
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV

# Definindo os valores que serão testados em SVM
c = np.array([1.0, 0.95, 1.05, 1.1, 1.2, 2, 0.9, 0.8])
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
polinomio = np.array([2, 3, 4, 5])
gamma = ['scale', 'auto']
valores_grid = {
    'C': c,
    'kernel': kernel,
    'degree': polinomio,
    'gamma': gamma
}

# Criando o modelo
modelo = SVC()

# Criando os grids
gridSVM = GridSearchCV(estimator=modelo, param_grid=valores_grid, cv=3, n_jobs=-1)

# Treinando o modelo
gridSVM.fit(x_normalizado, y)

# Resultado
print(f'Melhor valor constante: {gridSVM.best_estimator_.C}')
print(f'Melhor kernel: {gridSVM.best_estimator_.kernel}')
print(f'Melhor grau do polinomio: {gridSVM.best_estimator_.degree}')
print(f'Melhor gamma: {gridSVM.best_estimator_.gamma}')
print(f'Melhor score: {gridSVM.best_score_}')
```
